```python
df8=pandas.read_csv("supermarkets.csv")
df8
```

```python
df8["Address"]=df8["Address"]+", "+df8["City"]+", "+df8["Country"]
df8
```

```python
df8["Cordinates"]=df8["Address"].apply(nom.geocode)
```

```python
df8
```

```python
df8.Cordinates[0]
```

```python
df8.Cordinates[1]
```

```python
df8["Latitude"]=df8["Cordinates"].apply(lambda x: x.latitude if x != None else None)
df8["Longitude"]=df8["Cordinates"].apply(lambda x: x.longitude if x != None else None)
df8
```
